```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def saltPepperNoise(image):
    row, col, ch = image.shape
    s_vs_p = 0.5
    amount = 0.04
    noisy = np.copy(image)

    # Salt noise
    num_salt = int(np.ceil(amount * image.size * s_vs_p))
    coords = [np.random.randint(0, i-1, num_salt) for i in image.shape]
    coords = np.array(coords).T
    noisy[coords[:, 0], coords[:, 1], coords[:, 2]] = 1

    # Pepper noise
    num_pepper = int(np.ceil(amount * image.size * (1.0 - s_vs_p)))
    coords = [np.random.randint(0, i-1, num_pepper) for i in image.shape]
    coords = np.array(coords).T
    noisy[coords[:, 0], coords[:, 1], coords[:, 2]] = 0

    return noisy

# Resmi ykle ve baaryla yklenip yklenmediini kontrol et
img = cv2.imread("dag.jpg")
if img is None:
    print("Resim yklenemedi. Dosya adn ve yolunu kontrol et.")
else:
    img = img / 255.0  # Grnty normalize et

    noise_img = saltPepperNoise(img)
    noise_img = (noise_img * 255).astype(np.uint8)

    plt.imshow(noise_img)
    plt.title('Salt and Pepper Noise')
    plt.show()


```
