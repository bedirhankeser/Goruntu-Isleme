```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread("dag.jpg",0)

ret1, thresh1 = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

ret2, thresh2 = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Orjinal resmi göster
plt.subplot(131), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image')

# Basit thresholding sonucunu göster
plt.subplot(132), plt.imshow(thresh1, cmap='gray'), plt.title('Simple Thresholding')

# Otsu'nun thresholding sonucunu göster
plt.subplot(133), plt.imshow(thresh2, cmap='gray'), plt.title('Otsu Thresholding')

plt.show()
```
