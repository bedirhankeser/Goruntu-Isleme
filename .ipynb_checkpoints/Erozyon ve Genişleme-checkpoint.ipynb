```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('dag.jpg', 0)

kernel_erosion = np.ones((5, 5), np.uint8)

kernel_dilation = np.ones((5, 5), np.uint8)

img_erosion = cv2.erode(img, kernel_erosion, iterations=1)

img_dilation = cv2.dilate(img, kernel_dilation, iterations=1)

plt.figure(figsize=(12, 6))

plt.subplot(131), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(132), plt.imshow(img_erosion, cmap='gray'), plt.title('Erosion')
plt.subplot(133), plt.imshow(img_dilation, cmap='gray'), plt.title('Dilation')

plt.show()


```
